```python
import pandas as pd
import plotly
import plotly.express as px
from pathlib import Path
import xlwings as xw
# from ydata_profiling import ProfileReport
# from autoviz.AutoViz_Class import AutoViz_Class
```

```python
from pathlib import Path
import pandas as pd

input_path = Path('/content/input')
df = pd.DataFrame()

for file in input_path.iterdir():
  df = pd.concat([df, pd.read_excel(file)], ignore_index=True)

# df.head()
df['Country'].unique()
```

```python
df = df.groupby(by='Country').sum(numeric_only=True)[[" Sales", "Profit"]]
df
```

```python
# calculate profit
df['Profit Margin %'] = (df["Profit"] / df[' Sales']) * 100
df
```

```python
# Plot profit situation & export grouped dataframe to excel
fig = px.bar(
    df,
    x=df.index,
    y=" Sales",
    color="Profit Margin %",
    color_continuous_scale=["red", "yellow", "green"],
    template="plotly_white",
    title="<b>Profit Anlysis</b>",
)
fig.show()
fig.write_html('profit_analysis.html')
df.to_excel('profit_analysis.xlsx')
```
